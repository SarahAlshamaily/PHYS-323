```python
#Importing required packages

import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from scipy.integrate import quad
```

```python
#Function for integral

a = 2.632 #Normalization constant

def integ(z):
    integral = a*np.sqrt(2)*(np.sin(n*np.pi*z)*((2*z)+(0.3*z**2)-(2.3*z**3))*np.log((1+5*z)/2))
    return integral
```

```python
#Finding coefficients

coefficients = []

for n in range(1,16):
    c, d = quad(integ,0,1)
    coefficients.append(c)
    
print("The coefficients are: \n\n",coefficients,"\n\n")

coeff = [element*np.sqrt(2) for element in coefficients]

print("The new coefficients are: \n\n",coeff)
```

```python
m = np.linspace(0,1,100)

#Exact solution

exact = a*(2+0.3-2.3)*np.log((1+5)/2)

#Other approximated solutions for n=1, n=2, n=5, n=15

n1  = coeff[0]*np.sin(np.pi*m)
n2  = coeff[0]*np.sin(np.pi*m) + coeff[1]*np.sin(2*np.pi*m)
n5  = coeff[0]*np.sin(np.pi*m) + coeff[1]*np.sin(2*np.pi*m) + coeff[2]*np.sin(3*np.pi*m) + coeff[3]*np.sin(4*np.pi*m) + coeff[4]*np.sin(5*np.pi*m)
n15 = coeff[0]*np.sin(np.pi*m) + coeff[1]*np.sin(2*np.pi*m) + coeff[2]*np.sin(3*np.pi*m) + coeff[3]*np.sin(4*np.pi*m) + coeff[4]*np.sin(5*np.pi*m) + coeff[5]*np.sin(6*np.pi*m) + coeff[6]*np.sin(7*np.pi*m) + coeff[7]*np.sin(8*np.pi*m) + coeff[8]*np.sin(9*np.pi*m) + coeff[9]*np.sin(10*np.pi*m) + coeff[10]*np.sin(11*np.pi*m) + coeff[11]*np.sin(12*np.pi*m) + coeff[12]*np.sin(13*np.pi*m) + coeff[13]*np.sin(14*np.pi*m) + coeff[14]*np.sin(15*np.pi*m)

#Plotting the graph

plt.figure(1)
plt.figure(figsize=(10,10))
plt.plot(1,exact,'k',label='$exact$')
plt.plot(m,n1,'r',label='$n=1$',linestyle='--')
plt.plot(m,n2,'g',label='$n=2$',linestyle=':')
plt.plot(m,n5,'b',label='$n=5$',linestyle='-.')
plt.plot(m,n15,'m',label='$n=15$')
plt.title('Infinite Square Well Solutions')
plt.legend(loc='best')
plt.grid()
plt.show()
```
