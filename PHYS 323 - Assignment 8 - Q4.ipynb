```python
#Importing required packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define variables needed
a = 1.0 #a0
alpha = np.pi/2.0
alpha2 = np.pi/3.0
alpha3 = np.pi/4.0
alpha4 = np.pi/6.0
alpha5 = 0
y = 0
x,z = np.meshgrid(np.linspace(-10,10,100),np.linspace(-10,10,100))
r = np.absolute(np.sqrt(x**2 + y**2 + z**2))
prob = (((1.0/(10*np.pi*a**3)) * ((1 - (r/2.0*a)))**2)) + (((z*np.cos(alpha))/(20.0*np.pi*a**4))*(1-(r/(2.0*a)))) + (((z**2)/(160.0*np.pi*a**5)))*np.exp(-r/a)
prob2 = (((1.0/(10*np.pi*a**3)) * ((1 - (r/2.0*a)))**2)) + (((z*np.cos(alpha2))/(20.0*np.pi*a**4))*(1-(r/(2.0*a)))) + (((z**2)/(160.0*np.pi*a**5)))*np.exp(-r/a)
prob3 = (((1.0/(10*np.pi*a**3)) * ((1 - (r/2.0*a)))**2)) + (((z*np.cos(alpha3))/(20.0*np.pi*a**4))*(1-(r/(2.0*a)))) + (((z**2)/(160.0*np.pi*a**5)))*np.exp(-r/a)
prob4 = (((1.0/(10*np.pi*a**3)) * ((1 - (r/2.0*a)))**2)) + (((z*np.cos(alpha4))/(20.0*np.pi*a**4))*(1-(r/(2.0*a)))) + (((z**2)/(160.0*np.pi*a**5)))*np.exp(-r/a)
prob5 = (((1.0/(10*np.pi*a**3)) * ((1 - (r/2.0*a)))**2)) + (((z*np.cos(alpha5))/(20.0*np.pi*a**4))*(1-(r/(2.0*a)))) + (((z**2)/(160.0*np.pi*a**5)))*np.exp(-r/a)
```

```python
#Plotting required figure
plt.figure()
plt.imshow(prob, extent=[-10,10,-10,10], cmap='gray')
plt.colorbar()
plt.clim(0,0.0014)
plt.title('Density Function Plot for Alpha (pi/2)')
plt.xlabel('x')
plt.ylabel('z')
plt.show()

plt.figure()
plt.imshow(prob2, extent=[-10,10,-10,10], cmap='gray')
plt.colorbar()
plt.clim(0,0.0014)
plt.title('Density Function Plot for Alpha (pi/3)')
plt.xlabel('x')
plt.ylabel('z')
plt.show()

plt.figure()
plt.imshow(prob3, extent=[-10,10,-10,10], cmap='gray')
plt.colorbar()
plt.clim(0,0.0014)
plt.title('Density Function Plot for Alpha (pi/4)')
plt.xlabel('x')
plt.ylabel('z')
plt.show()

plt.figure()
plt.imshow(prob4, extent=[-10,10,-10,10], cmap='gray')
plt.colorbar()
plt.clim(0,0.0014)
plt.title('Density Function Plot for Alpha (pi/6)')
plt.xlabel('x')
plt.ylabel('z')
plt.show()

plt.figure()
plt.imshow(prob5, extent=[-10,10,-10,10], cmap='gray')
plt.colorbar()
plt.clim(0,0.0014)
plt.title('Density Function Plot for Alpha (0)')
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```
