```python
#Importing required packages

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#Defining variables and functions required

a = 1
z0 = 1

x = np.linspace(-10,10,1000)
z = np.linspace(-10,10,1000)
y = 0
X, Z = np.meshgrid(x, z)
r = np.sqrt(X**2 + y**2 + Z**2)

func200 = (1/np.sqrt(np.pi))*((z0/(2*a))**(3/2))*(1-(z0*r)/(2*a))*np.exp(-(z0*r)/(2*a))
func210 = (1/(2*np.sqrt(np.pi)))*((z0/(2*a))**(3/2))*((z0*r)/a)*np.exp(-(z0*r)/(2*a))*(Z/r)
```

```python
#Plotting the density plots using a for loop that iterates through different alpha values

for alpha in [0,0.1*np.pi,0.2*np.pi,0.3*np.pi,0.4*np.pi,0.5*np.pi]:
    wavefunction = np.sqrt(3/5)*func200 + np.exp(1j*alpha)*np.sqrt(2/5)*func210
    probdens = ((np.conj(wavefunction)*wavefunction)).real
    
    plt.figure(figsize=(5,5))
    plt.pcolormesh(X, Z, probdens, vmin=0, vmax=0.0014, cmap="gray")
    plt.xlabel("x")
    plt.ylabel("z")
    plt.colorbar()
    plt.title('Probability Density Function')
    plt.show()
```

**Part (b)**

In this question, alpha is the relative phase. Therefore, we can expect to get a different "weighting" of the state 
|2,1,0> that can change the density plots each time we use a different alpha value, because that state is dependent on the value of alpha. So increasing the relative phase will also increase the probability density in +z.
