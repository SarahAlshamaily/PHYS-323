```python
#Importing required packages
import numpy as np
import scipy.constants as c
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Introducing required constants and values
pi = np.pi
m = c.m_e
h = c.hbar
w = 2*pi*5*(10**15)

x0 = -2*(10**-10) #Initial x limit
x1 = 2*(10**-10)  #Final   x limit
y0 = 0            #Initial y limit
y1 = 10**-17      #Final   y limiy
n = 10000
x = np.linspace(x0,x1,nx)
```

```python
#Required psi functions
psi0 = (((m*w)/(pi*h))**(1/4))*np.exp(-(m*w*(x**2))/(2*h))
psi1 = (((m*w)/(pi*h))**(1/4))*(((2*m*w)/h)**(1/2))*x*np.exp(-(m*w*(x**2))/(2*h))
psi2 = (1/(h*np.sqrt(2)))*(((m*w)/(pi*h))**(1/4))*np.exp(-(m*w*(x**2))/(2*h))*(2*m*w*(x**2)-(h**2))

#Required energies
E0 = (0+(1/2))*h*w
E1 = (1+(1/2))*h*w
E2 = (2+(1/2))*h*w
```

```python
l = np.ones(n) #Filling array with ones

#Required waves
w0 = E0 + (10**(-6))*E2*psi0
w1 = E1 + (10**(-6))*E2*psi1
w2 = E2 + (10**(-6))*E2*psi2

#Equation for simple harmonic oscillator
simharosc = (1/2)*m*(w**2)*(x**2)
```

```python
#Producing required plot
plt.figure(figsize=[8,10])
plt.title("Simple Harmonic Oscillator (SHO)")
plt.xlabel("x (meters)")
plt.ylabel("Potential, V(x) (Joules)")
plt.plot(x,simharosc,'k',label="$Simple Harmonic Oscillator\ V$")
plt.plot(x,E0*l,'--r',label="$E_O$")
plt.plot(x,w0,'r',label="$\psi_0$")
plt.plot(x,E1*l,'--g',label="$E_1$")
plt.plot(x,w1,'g',label="$\psi_1$")
plt.plot(x,E2*l,'--b',label="$E_2$")
plt.plot(x,w2,'b',label="$\psi_2$")
plt.xlim(x0,x1)
plt.ylim(y0,y1)
plt.grid()
plt.legend(loc="best")
plt.savefig("SHO.png")
plt.show()
```
